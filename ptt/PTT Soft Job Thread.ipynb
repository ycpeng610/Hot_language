```python
import requests
from bs4 import BeautifulSoup
from collections import Counter
import re
import time
import threading

wc = Counter()
wc["C"] = 0
wc["C++"] = 0
wc["C#"] = 0
wc["PYTHON"] = 0
wc["JAVA"] = 0
wc["JAVASCRIPT"] = 0
wc["PHP"] = 0
wc["HTML"] = 0
wc["SQL"] = 0
wc["CSS"] = 0
wc["R"] = 0
wc["BASH"] = 0
wc["RUBY"] = 0
wc["PERL"] = 0
wc["SCALA"] = 0
wc["SWIFT"] = 0
wc["GO"] = 0
wc["DELPHI"] = 0
wc["TYPESCRIPT"] = 0
wc
```

```python
def getLink(page):
    print('thread' + str(page) +'start')

    try:
        res = requests.get(HOST + "/bbs/Soft_Job/index{}.html".format(page),headers = headers)
        soup = BeautifulSoup(res.text,'lxml')
        links = soup.select('div.title > a')
        for link in links: #link 是每一篇文
            if '徵才' in link.text:
#                 print(link.text)
                each_link = HOST + link['href']
#                 print(each_link)
#                     ptt(each_link)
                pttword = pttThread(each_link)
                pttword.start()
                pttword.join()
    except Exception as e:
        print(e,link['href'])
```

```python
class getLinkThread(threading.Thread):
    def __init__(self,page):
        threading.Thread.__init__(self)
        self.page = page
    def run(self):
        getLink(self.page)
```

```python
def ptt(each_link):
    
    res = requests.get(each_link,headers = headers)
    soup = BeautifulSoup(res.text,'lxml')

    content = soup.select_one('#main-content') #內文
    
    words = list(set(re.findall('[A-Z]+[+#]*' , content.text, re.IGNORECASE)))#不處理大小寫,set處理重複問題
    
    for language in words:
        global wc
        if language.upper() in wc.keys():
            wc[language.upper()] +=1
```

```python
class pttThread(threading.Thread):
    def __init__(self,each_link):
        threading.Thread.__init__(self)
        self.each_link = each_link
    def run(self):
        ptt(self.each_link)
```

```python
url = 'https://www.ptt.cc/bbs/Soft_Job/index1247.html' #ptt最新頁面

HOST = 'https://www.ptt.cc'
headers = {"cookie":"over18=1;"}

#換頁
res = requests.get(url,headers = headers)
soup = BeautifulSoup(res.text,'lxml')

buttons = soup.select('a.btn.wide')
threads = []

total_page = int(buttons[1]['href'].split('index')[1].split('.html')[0])+1
page_to_crawl = 1238
for page in range(total_page, total_page - page_to_crawl, -1): #單頁
    Thread = getLinkThread(page)
    threads.append(Thread)
for i in threads:
    i.start()
    time.sleep(0.1)
for i in threads:
    i.join()




            

# if page % 50 == 0:
#         print('[Debug] current page is {}'.format(page))
wc.most_common()
```
